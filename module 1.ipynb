# 23. Для какого режиссера зима – самое продуктивное время года? 

```python
answers['23'] = 'Clint Eastwood'
```

```python
from pandasql import *
from pandasql import sqldf, load_meat, load_births
pysqldf = lambda q: sqldf(q, globals())
```

```python
pysqldf = lambda q: sqldf(q, globals()) #инициалация pandasql    globals()- глобальные переменные системы
query = """select  director,release_date,
case when (release_date like '1/%') OR  (release_date like '2/%')OR release_date like '12/%' then  1
 when (release_date like '3/%') OR  (release_date like '4/%')OR release_date like '5/%' then  2
 when (release_date like '6/%') OR  (release_date like '7/%')OR release_date like '8/%' then  3
 when (release_date like '9/%') OR  (release_date like '10/%')OR release_date like '11/%' then  4
END AS SEASON
     FROM
       data """ #limit 5 = head(5)
data_tmp =pysqldf(query) # execution
```

```python
directors_set=set()

for element in data_tmp['director'].unique():
    for element2 in element.split('|'):
        directors_set.add(element2)
```

```python
data_tmp.info()
```

```python
winter_directors=[]
for element in directors_set:
    data_tmp2=data_tmp[(data_tmp.director.str.contains(element, na=False))]
    winter=data_tmp2[(data_tmp2['SEASON']==1)]
    spring=data_tmp2[(data_tmp2['SEASON']==2)]
    summer=data_tmp2[(data_tmp2['SEASON']==3)]
    fall=data_tmp2[(data_tmp['SEASON']==4)]
    print(len(fall))
    if len(winter)>max(len(spring),len(summer),len(fall)):    
        winter_directors.append(element)
```
